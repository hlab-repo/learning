## Softmax
The softmax function converts logits into a probability vector, where every element of the vector >= 0 and <= 1, and where the sum of the entire vector is 1. This function generalizes logistic regression to multiple dimensions and thus usesd the natrual exponential function.

$softmax(z_i) = \frac{e^{z_i}}{\sum_{j=1}^{n} e^{z_j}}$

```python
z = np.array([-1.2, 0.3, 2.8, -3.1])
softmax = np.exp(z) / np.sum(np.exp(z))
softmax                                  # 0 <= item <= 1 for every item, and the sum of all items == 1.
```

## Temperature
The temperature (a term from statistical physics) is a constant that controls the distribution of the probability mass. High temperature spreads the mass uniformly, whereas low temperature puts most of the mass on the most probable item(s).

Screen Shot 2021-04-05 at 7.50.52 AM.png

$softmax\_with\_temp(z_i) = \frac{e^{z_i / T}}{\sum_{j=1}^{n} e^{z_j / T}}$

```python
z = np.array([-1.2, 0.3, 2.8, -3.1])  # same as above
temp = 1.2
softmax = np.exp(z / temp) / np.sum(np.exp(z / temp))
softmax 
```

## KL Divergence
The Kullback-Leibler (KL) divergence measures the distance between two probability distributions $P$ and $Q$. For discrete distributions, the KL divergence is defined as follows:

$KL(p||q) = \sum_{k=1}^{K} p_k log(\frac{p_k}{q_k})$

```python
# different p and q means there should be some distance between p and q
p = np.array([.2, .3, .4, .1])             # probability distribution 1
q = np.array([.1, .2, .4, .3])             # probability distirbution 2
kl_divergence = np.sum(p * np.log(p / q))
kl_divergence
```

```python
# if p and q are the same, then there should be 0 distance
p = np.array([.1, .15, .6, .15])             # probability distribution 1
q = np.array([.1, .15, .6, .15])             # probability distibution 2
kl_divergence = np.sum(p * np.log(p / q))
kl_divergence
```
